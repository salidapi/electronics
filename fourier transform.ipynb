```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# Prkufizimi i domenit t kohs
T = 1.0  # Periudha totale e kohs
N = 500  # Numri i mostrave
t = np.linspace(0, T, N, endpoint=False)

# Prkufizimi i nj funksioni: Shuma e sinusoideve
f = np.sin(2 * np.pi * 5 * t) + 0.8 * np.sin(2 * np.pi * 10 * t)

# Llogaritja e Transformimit t Fourier-it
F = fft(f)
frequencies = fftfreq(N, T/N)

# Grafiku i sinjalit n domenin e kohs
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t, f)
plt.title("Sinjali n Domenin e Kohs")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")

# Grafiku i spektrit t frekuencs
plt.subplot(1, 2, 2)
plt.plot(frequencies[:N//2], np.abs(F[:N//2]))
plt.title("Spektri i Frekuencs (Transformimi i Fourier-it)")
plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Madhsia")
plt.xlim(0,30)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

#Domeni i kohes
T= 1.0
N=500
t=np.linspace(0, T, N, endpoint=False)
 #Funksioni
f=np.sin(2*np.pi*np.exp(5*t))+ 0.8 * np.sin(2 * np.pi * 10 * t)

# Llogaritja e Transformimit t Fourier-it
F = fft(f)
frequencies = fftfreq(N, T/N)

# Grafiku i sinjalit n domenin e kohs
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t, f)
plt.title("Sinjali n Domenin e Kohs")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")

# Grafiku i spektrit t frekuencs
plt.subplot(1, 2, 2)
plt.plot(frequencies[:N//2], np.abs(F[:N//2]))
plt.title("Spektri i Frekuencs (Transformimi i Fourier-it)")
plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Madhsia")
plt.xlim(0,30)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrat e sinjalit
A = 2
f = 5  # Hz
T = 1 / f  # Periudha
t = np.linspace(0, 2*T, 1000)  # 2 periudha

# Gjenerimi i sinjalit
signal = A * np.cos(2 * np.pi * f * t)

# Llogaritja e mesatares kohore dhe fuqis
mean_signal = np.mean(signal)
mean_power = np.mean(signal**2)

# Vizualizimi i sinjalit
plt.plot(t, signal, label="Sinjali")
plt.axhline(mean_signal, color='r', linestyle='--', label=f"Mesatarja = {mean_signal:.2f}")
plt.title("Mesatarja Kohore e nj Sinjali Sinusoidal")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")
plt.legend()
plt.show()

print(f"Mesatarja e sinjalit: {mean_signal:.4f}")
print(f"Mesatarja e fuqis: {mean_power:.4f} (pritet A^2/2 = {A**2/2:.2f})")
```
